**What's data wrangling?**
Data wrangling is a process of transforming your data from a raw format to a format that is more appropriate for downstream analysis. You'll have to assess the quality of the current raw state of the data, and make decisions about how you'll go about improving its quality. 

The time you'll spend in data wrangling will depend on the quality of the data you get. For me, this is often the longest process. When working with data that are given to me, I spend time exploring the data and the metadata (if available), e.g., size of data and data types present. Then, I move on to assess how data are formatted as I often have to deal with spaces where there shouldn't be any, formatting inconsistencies, missing-values, etc. After all of that, I finally proceed to preprocessing it for my downstream analysis.

To demonstrate how to do data wrangling, I am using the Gapmider data from one of [The Carpentries](http://swcarpentry.github.io/python-novice-gapminder/) Workshop [I taught in the past](https://github.com/nathaliagg/my_carpentries_workshops).

## Rename column names

There are many ways to do this, and it will also depend on how many column you have and which ones you want to rename.

In this case we have 38 columns, and let's say I want to change columns name to the following: `Continent`, `Country`, `GDPpercap_*`, `LifeExp_*`, `Pop_*`. I can write a `for loop` to iterate over column names and to deal with all of them at once.

In order to maintain the raw format, I'll first create a copy of `df_all` as `df_mod`, in which *mod* stands for modified. 

```python
df_mod = df_all.copy()
```

```python
# rename all columns:
for col in list(df_mod.columns):
    # print(col) ## always use print statements while building your for loop
    if col.startswith('c'):
        # print(col.capitalize())
        df_mod.rename(columns={col: col.capitalize()}, inplace=True)
    elif col.startswith('gdp'):
        # first get the year:
        year = col.split('_')[-1]
        new_name = 'GDPpercap_' + year
        # print(new_name)
        df_mod.rename(columns={col: new_name}, inplace=True)
    elif col.startswith('l'):
        year = col.split('_')[-1]
        new_name = 'LifeExp_' + year
        # print(new_name)
        df_mod.rename(columns={col: new_name}, inplace=True)
    else:
        # print(col.capitalize())
        df_mod.rename(columns={col: col.capitalize()}, inplace=True)
```

## Rename rows

For example, let's rename `Brazil` to `Brasil` because Brasil it's how Brazilians write their country name ;)

I'll demonstrate two ways:
- using `iterrows()`
- accessing one specific value using `at[]`

```python
# make a copy
df_mod2 = df_mod.copy()

# test to see if Brazil is there
df_mod2[df_mod2['Country']=='Brazil']
```

### using `iterrows()`

This is useful when you have quite a few things to rename/change.

```python
# iterate through rows
for index,row in df_mod2.iterrows():
    # print(index, row)
    # print(index, row['Continent'])
    if row['Country'] == 'Brazil':
        # print(index, row['Country'])
        df_mod2.loc[54, 'Country'] = 'Brasil' # this will replace Brazil to Brasil in the df, hence the df copy
    # if you have more values to change, you could continue the for loop using the appropriate conditionals

# test if the modification was successful, does Brazil still exist?
df_mod2[df_mod2['Country']=='Brazil']
```

```python
# test if the modification was successful
df_mod2[df_mod2['Country']=='Brasil']
```

### accessing one specific value using `at[]`

```python
# re-write the copy df
df_mod2 = df_mod.copy()

# test to see if Brazil is there
df_mod2[df_mod2['Country']=='Brazil']
```

```python
# use to replace Brazil to Brazil as at[index, column] = value
df_mod2.at[54,'Country'] = 'Brasil'

# test if the modification was successful
df_mod2[df_mod2['Country']=='Brasil']
```

## Pivot tables from wide to long

Let's use `melt`.

```python
# subset the df to make it easier to visualize
# subset Continent, Country and GDPpercap_*
subset_df = df_mod.iloc[:,:14]
subset_df.head()
```

```python
# 1- let's get a list of columns we want to collapse
target_columns = [
    col for col in subset_df.columns.values if col.startswith('GDP')
]

df_long = pd.melt(subset_df, 
        id_vars=['Continent', 'Country'], 
        value_vars=target_columns,
        var_name = 'Parameter',
        value_name = 'Value')

df_long.head()
```

## Pivot tables from long to wide

Let's use `pivot_table` because it can handle duplicate values in the index.

```python
df_wide = pd.pivot_table(df_long, 
               index = ['Continent', 'Country'],
               values = 'Value', 
               columns='Parameter').reset_index()
df_wide.head()
```

## Split columns

Let's use `df_long` to split the Parameter column into Parameter and Year:

```python
df_long['Year'] = df_long['Parameter'].str.split('_').str[-1]
df_long['Parameter'] = df_long['Parameter'].str.split('_').str[0]
df_long.head()
```

## Join columns

Or we could concat two columns, Parameters and Year, into one. There are a few ways to do so:

```python
# 1- hard code
df_long['Parameter'] + '_' + df_long['Year']
```

```python
# 2- using apply and lambda
df_long[['Parameter','Year']].apply(lambda x: "_".join(x), axis=1)
```

```python
# 3- using aggregate
df_long[['Parameter','Year']].agg('_'.join, axis=1)
```

```python
# choose one of the methods 
# apply the df and drop Year
df_long['Parameter']= df_long[['Parameter','Year']].apply(lambda x: "_".join(x), axis=1)
df_long.drop('Year', axis=1, inplace=True)
df_long
```
